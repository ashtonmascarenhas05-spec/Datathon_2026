```python
## Converting CrimeRegisteredDate to datetime format
CaseMaster['CrimeRegisteredDate'] = pd.to_datetime(CaseMaster['CrimeRegisteredDate'], errors='coerce')
```

```python
## Converting IncidentFromDate, IncidentToDate,InfoReceivedPSDate to datetime format
CaseMaster['IncidentFromDate'] = pd.to_datetime(CaseMaster['IncidentFromDate'], errors='coerce')
CaseMaster['IncidentToDate'] = pd.to_datetime(CaseMaster['IncidentToDate'], errors='coerce')
CaseMaster['InfoReceivedPSDate'] = pd.to_datetime(CaseMaster['InfoReceivedPSDate'], errors='coerce')
```

```python
CaseMaster.info()
```

```python
## Checking if the IncidentFromDate is before the IncidentToDate in CaseMaster dataframe
for index, row in CaseMaster.iterrows():
    incident_from_date = row['IncidentFromDate']
    incident_to_date = row['IncidentToDate']
    
    if pd.notnull(incident_from_date) and pd.notnull(incident_to_date) and incident_from_date > incident_to_date:
        print(f"Warning: IncidentFromDate {incident_from_date} is after IncidentToDate {incident_to_date} for CaseMasterID {row['CaseMasterID']}")
```

```python
## Checking if the CrimeRegisteredDate is after one day the IncidentFromDate and IncidentToDate
for index, row in CaseMaster.iterrows():
    crime_date = row['CrimeRegisteredDate']
    incident_from_date = row['IncidentFromDate']
    incident_to_date = row['IncidentToDate']
    one_day_after_incident_from = incident_from_date + pd.Timedelta(days=1)
    one_day_after_incident_to = incident_to_date + pd.Timedelta(days=1)

    if pd.notnull(crime_date) and pd.notnull(incident_from_date) and crime_date < one_day_after_incident_from:
        print(f"Warning: CrimeRegisteredDate {crime_date} is before one day after IncidentFromDate {one_day_after_incident_from} for CaseMasterID {row['CaseMasterID']}")

    if pd.notnull(crime_date) and pd.notnull(incident_to_date) and crime_date < one_day_after_incident_to:
        print(f"Warning: CrimeRegisteredDate {crime_date} is before one day after IncidentToDate {one_day_after_incident_to} for CaseMasterID {row['CaseMasterID']}")
```

```python
## Just to check if the ArrestSurrender dataframe's ArrestSurrenderDate is after the CrimeRegisteredDate in CaseMaster dataframe
for index, row in ArrestSurrender.iterrows():
    case_id = row['CaseMasterID']
    arrest_date = row['ArrestSurrenderDate']
    crime_date = CaseMaster.loc[CaseMaster['CaseMasterID'] == case_id, 'CrimeRegisteredDate'].values[0]
    one_day_after_crime_date = crime_date + pd.Timedelta(days=1)

    if pd.notnull(arrest_date) and pd.notnull(crime_date) and arrest_date < one_day_after_crime_date:
        print(f"Warning: ArrestSurrenderDate {arrest_date} is before one day after CrimeRegisteredDate {one_day_after_crime_date} for CaseMasterID {case_id}")
```

```python
## Just to check if the ArrestSurrender dataframe's ArrestSurrenderDate is after one day of the IncidentToDate in CaseMaster dataframe
for index, row in ArrestSurrender.iterrows():
    case_id = row['CaseMasterID']
    arrest_date = row['ArrestSurrenderDate']
    incident_to_date = CaseMaster.loc[CaseMaster['CaseMasterID'] == case_id, 'IncidentToDate'].values[0]
    one_day_after_incident = incident_to_date + pd.Timedelta(days=1)

    if pd.notnull(arrest_date) and pd.notnull(incident_to_date) and arrest_date < one_day_after_incident:
        print(f"Warning: ArrestSurrenderDate {arrest_date} is before one day after IncidentToDate {one_day_after_incident} for CaseMasterID {case_id}")
```
