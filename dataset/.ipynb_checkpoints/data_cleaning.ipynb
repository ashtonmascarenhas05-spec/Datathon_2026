```python
import pandas as pd

def find_primary_keys(df):
    """
    Find columns that could be Primary Keys.
    Conditions:
    1. No missing values
    2. All values are unique
    """
    primary_keys = []

    for col in df.columns:
        if df[col].isnull().sum() == 0 and df[col].is_unique:
            primary_keys.append(col)

    return primary_keys
```

```python
find_primary_keys(employee)
```
